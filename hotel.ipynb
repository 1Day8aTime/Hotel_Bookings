DATA ANALYSIS AND VISUALISATION BEGINS

```python
# PERCENTAGE OF RESERVATION CANCELLED
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

```

SHOWS ~62% - NOT CANCELLED

(38 is very high)
lets visualise this to understand more

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

#plot a BAR GRAPH to compare the cancelled and not cancelled 
plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(['Not Cancelled','Cancelled'], df['is_canceled'].value_counts(), edgecolor = 'k' , width = 0.5)
plt.show()
```

```python
# PLOT TO DETERMINE WHICH TYPE OF HOTEL HAS HOW MUCH CANCELLATIONS
# PLOT BAR GRAPH USING SEABORN
plt.figure(figsize = (8,4))
axl = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df)
legend_labels,_ = axl. get_legend_handles_labels()
axl.legend(loc='upper right')
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.legend(['not_cancelled','cancelled'])
plt.show();
```

Cancellation ratio is high in city hotel as compared to resort hotel as seen from above plot

```python
# lets see % of cancellation seperately in resort hotel 

resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

~28 % cancelled in resort hotels

```python
# lets see % of cancellation seperately in city hotel

city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

~42% cancellation in city hotel WHICH IS HIGH

WE FIND THE AVG ADR FOR RESORT AND CITY HOTELS ON RESERVATIN STATUS DATE

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

PLOT THE AVG ADR ON DIFFERENT RESERVATION STATUS DATE FOR CITY AND RESORT HOTELS


```python
plt.figure( figsize = (20,8))
plt.title('Average Daily rate in city and resort hotels', fontsize = 20)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
plt.figure( figsize = (20,5))
plt.title('Average Daily rate in city and resort hotels', fontsize = 30)
plt.plot( resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

ANALYSE CANCELLATIONS IN DIFFERENT MONTHS

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,5))
ax = sns.countplot(x = 'month', hue = 'is_canceled', data = df)
legend_labels,_ = ax. get_legend_handles_labels()
ax.legend(loc = 'upper right')
plt.title('Reservation status per month')
plt.xlabel('Month')
plt.ylabel('Number of reservations')
plt.legend(['not cancelled','cancelled'])
plt.show()
```

AUGUST - HIGHEST RESEVATION AND LOWEST CANCELLATION 
JAN - LOWEST RESERVATIO AND HIGHEST CANCELLATION

WHAT IS THE REASON FOR LOW CANCELLATION ??
LOW ADR - SO HIGH - NOT CANCELLED ?


```python
#FIND  -RELATION OF ADR EACH MONTH WITH CANCELLATION
plt.figure(figsize = (15,5))
plt.title('ADR per Month', fontsize = 30)
dt = df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index()
sns.barplot(x = 'month', y = 'adr', data = dt)
plt.legend(fontsize = 20)
plt.show()
```

SHOWS HIGH ADR MEANS HIGH CANCEL

```python
# COUNTRY - WISE ANALYSIS FOR TOP 10 COUNTRIES
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (6,6))
plt.title('Top 10 countries with reservation cancelled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

The country with the code 'PRT' amounts to very high cancellations, which are around 70%.

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

```

```python
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate', fontsize = 20)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
plt.legend(fontsize = 15)
```
